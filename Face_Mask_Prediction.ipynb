# Load Model

```python
from tensorflow.keras.models import load_model
classifier = load_model(filepath='/content/drive/MyDrive/MSIB/BISA AI/Computer Vision/mask_classifier.keras')
```

# Crop Image

```python
import cv2
import os
from google.colab import files
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Unggah gambar
uploaded = files.upload()

# Ambil path gambar yang diunggah
image_path = list(uploaded.keys())[0]

# Load Haar Cascade Classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def crop_faces(input_image_path, output_dir): # Change input_dir to input_image_path
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Instead of iterating through a directory, process the single image
    img = cv2.imread(input_image_path) # Use input_image_path directly
    if img is None:
        print(f"Error: Could not read image at {input_image_path}")
        return

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        cropped_face = img[y:y+h, x:x+w]
        # Use the original image name when saving
        save_path = os.path.join(output_dir, os.path.basename(input_image_path))
        cv2.imwrite(save_path, cropped_face)
        print(f"Cropped and saved: {save_path}")

# Example usage, pass the actual image path to the function
crop_faces(image_path, 'foto_cropped')
```

# Prediksi Data Baru


```python
image_path = '/content/foto_cropped/1650947517923.jpg'
# Load dan preprocess gambar
img = cv2.imread(image_path)  # Load the image first
img = cv2.resize(img, (128, 128))  # Resize the image to (128, 128)
img_array = np.array([img])/255.0

# Calculate probabilities using the model
probabilities = classifier.predict(np.expand_dims(img, axis=0)) # This line is added to calculate probabilities

class_indices = {0: 'with_mask', 1: 'improper_mask', 2: 'without_mask'}
classes = ['mask_weared_incorrect', 'with_mask', 'without_mask']
# Menampilkan probabilitas untuk setiap kelas
for i, probability in enumerate(probabilities[0]):
    print(f"Probabilitas kelas {classes[i]}: {probability:.4f}")

# Melakukan prediksi
prediction = np.argmax(probabilities)
predicted_label = classes[prediction]

# Tampilkan gambar dan hasil prediksi
plt.imshow(image.load_img(image_path)) # This line might need adjustment if you want to display the resized image
plt.axis('off')
plt.title(f"pred: {predicted_label}")
plt.show()
```
